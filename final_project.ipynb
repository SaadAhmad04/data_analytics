Modelling

The highest sale is in the West so the highest Profit is also in the West but there is an issue with the dataset that the sale is lowest in the South while the profit is lowest in the Central Region

```python
df.groupby("Region")["Sales"].sum().plot.bar()
```

```python
df.groupby("Region")["Profit"].sum().plot.bar()
```

```python
df.groupby("State")["Profit"].sum().plot.bar()
```

The negative profit shows that there is a loss in some state despite of being some sales. So the company should not sell their products in the states where there is a loss or should think something different for these states.

```python
df.groupby("Category")["Sales"].sum().plot.pie(autopct="%1.0f%%") # type: ignore
```

The above pie chart shows that the highest number of sales is in the Technology category

```python
#The company is in loss because of the Bookcases , Supplies and Tables
df.groupby("Sub-Category")["Profit"].sum().plot.bar()
```

```python
#from the above graph , we can see that the company is getting the highest profit in Copiers also they sell the lowest quantity of Copiers
df.groupby("Sub-Category")["Quantity"].sum().plot.bar()
```

```python
#The company is giving highest discount in the Binders
df.groupby("Sub-Category")["Discount"].sum().plot.bar() 
```

```python
#the company is getting the highest profit from the Consumers i.e 134007.4413
df.groupby("Segment")["Profit"].sum().plot.line()
```
